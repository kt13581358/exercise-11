```python
class IceCreamShop:
    MAX_STOCK = 200  
    
    def __init__(self):
        self.ice_creams = {
            "شکلاتی": {"weight": 0.1, "price": 10, "count": 0},
            "میوه‌ای": {"weight": 0.055, "price": 7, "count": 0},
            "موزیکال": {"weight": 0.2, "price": 15, "count": 0, "song": "🎵 آهنگ ویژه بستنی موزیکال 🎵"},
            "یخی": {"weight": 0.04, "price": 3, "count": 0}
        }
        self.total_weight = 0
    
    def add_ice(self, name, amount):
        if name not in self.ice_creams:
            print("این نوع بستنی وجود ندارد!")
            return
        
        added_weight = self.ice_creams[name]["weight"] * amount
        if self.total_weight + added_weight > self.MAX_STOCK:
            print(f"ظرفیت انبار کافی نیست! (ظرفیت باقیمانده: {self.MAX_STOCK - self.total_weight:.2f} کیلوگرم)")
            return
        
        self.ice_creams[name]["count"] += amount
        self.total_weight += added_weight
        print(f"{amount} عدد بستنی {name} به انبار اضافه شد.")
    
    def sell_ice(self, name, amount, play_song=False):
        if name not in self.ice_creams:
            print("این نوع بستنی وجود ندارد!")
            return
        
        if name == "موزیکال" and amount % 2 != 0:
            print("بستنی موزیکال فقط به صورت جفتی فروخته می‌شود!")
            return
        
        if self.ice_creams[name]["count"] < amount:
            print(f"موجودی بستنی {name} کافی نیست! (موجودی: {self.ice_creams[name]['count']})")
            return
        
        sold_weight = self.ice_creams[name]["weight"] * amount
        total_price = self.ice_creams[name]["price"] * amount
        profit = total_price * 0.2
        
        self.ice_creams[name]["count"] -= amount
        self.total_weight -= sold_weight
        
        print(f"\nفروش موفق: {amount} عدد بستنی {name}")
        print(f"وزن فروخته شده: {sold_weight:.3f} کیلوگرم")
        print(f"قیمت کل: {total_price} هزار تومان")
        print(f"سود فروش: {profit} هزار تومان")
        
        if name == "موزیکال" and play_song:
            print(self.ice_creams[name]["song"])
    
    def show_stock(self):
        print("\nوضعیت انبار:")
        print(f"موجودی کل: {self.total_weight:.3f}/{self.MAX_STOCK} کیلوگرم")
        for name, details in self.ice_creams.items():
            print(f"{name}: {details['count']} عدد")
```

```python
shop = IceCreamShop()

shop.add_ice("شکلاتی", 100)
shop.add_ice("میوه‌ای", 150)
shop.add_ice("موزیکال", 50)  
shop.add_ice("یخی", 200)

shop.show_stock()

shop.sell_ice("شکلاتی", 5)
shop.sell_ice("موزیکال", 3)  
shop.sell_ice("موزیکال", 4, play_song=True)  
shop.sell_ice("یخی", 250)  

shop.show_stock()
```
